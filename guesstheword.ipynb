```python
import random
wordlist = ["Apple", "banana", "blueberry", "kiwi", "baboon"]
chosen_word = random.choice(wordlist).lower()
print(chosen_word)
lives = len(chosen_word)
inputlength = len(chosen_word)
print(lives)
display = ["_"] * len(chosen_word)
gameover = False
collectedword=[]
incorrectwords=[]
while not gameover:
    guess = input("enter a letter: ").lower()
    for i, letter in enumerate(chosen_word):
        if letter == guess:
            display[i] = guess
            collectedword.append(guess)
        
    if guess not in collectedword:
        lives-=1  
        if guess in incorrectwords:
            print(f"{guess} is already guessed and is incorrect")
        incorrectwords.append(guess) 
    
        print(f"you have {lives} lives out of {inputlength} left. Choose carefully")

    if lives==0:
        ##print("in lives loop")
        print("You are our of lives.. Please try again")
        break
    print("".join(display))
    if "_" not in display:
        gameover = True
        print("You WON!!!")
```
