    Nested Dictionaries

```python
student_list = {
    "student1":{"name":"Uma", "age":"38"},
    "student2":{"name":"Pavan", "age":"25"}
}
print(student_list)
print(student_list["student1"]["name"])
```

```python
dict1={'a':'1', 'b':'2' }
dict2={'b':'3', 'd':'4' }
merged_dict = {**dict1, **dict2} ##
dict1.update(dict2)
print(dict1)  # Output: {'a': '1', 'b': '3', 'd': '4'}
print(merged_dict)  # Output: {'a': '1', 'b': '2', 'c': '3', 'd': '4'}
```

```python
print(student_list)
for student1, student2 in student_list.items():
    print(f"Student Details for {student1}:")
    for key, value in student2.items():
        print(f"key is {key},\n value is {value}")
```

```python
numbers=[1,2,3,3,4,5,3,6,7,1,2,5,8,0,8]
## get the count of each number in the list and store it in a dictionary
count_dict={}
for num in numbers:
    if num in count_dict:
        count_dict[num] += 1
    else:
        count_dict[num] = 1
print(count_dict)
```

Numbers and their squares in dictionaries

```python
number_squres = {x:x**2 for x in range(1,11)}
print(number_squres)
print(f"Value of key 5: {number_squres[5]}")
print(f"Keys: {list(number_squres.keys())}")
print(number_squres.keys())
```

```python
##add a new key value pair to the dictionary
number_squres[11]=121
print(number_squres)
##remove a key value pair from the dictionary
del number_squres[11]
print(number_squres)

remove_key=number_squres.pop(10)
print(remove_key)
```

```python
for key, value in number_squres.items():
    print(f"key is {key}, value is {value}")
```
