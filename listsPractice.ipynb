```python
def min_in_tuple(tpl):
    return min(tpl)

def max_in_tuple(tpl):
    return max(tpl)

def sum_of_tuple(tpl):
    return sum(tpl)

sample_tpl = (8, 9, 10, 100)
print(f"Minimum: {min_in_tuple(sample_tpl)}")
print(f"Maximum: {max_in_tuple(sample_tpl)}")
print(f"Sum: {sum_of_tuple(sample_tpl)}")
```

```python
myset={1,2,3,4,5}
print(myset)
```

```python
myset.add(6)
print(myset)
##myset.remove(3)
print(myset)
removed_element = myset.pop() ##pop removes and returns an arbitrary element from the set
print(f"Removed element: {removed_element}")
myset.update([1, 2, 3, 4, 5, 6])
print(myset)
myset.discard(22) ##discard does not raise an error if the element is not found, while remove would raise a KeyError
print(myset)
```

```python
print(3 in myset) ##checking if 3 is in the set
print(10 in myset) ##checking if 10 is in the set
```
