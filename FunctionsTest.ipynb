POSITIONAL ARGS

```python
def print_numbers(*args): ## *args is used to pass a variable number of arguments to a function. It allows you to pass any number of positional arguments to the function, and they will be accessible as a tuple within the function.
    for num in args:
        print(num)

print_numbers(1, 2, 3, 4, 5, 'Pavan')
## *args is used to pass a variable number of arguments to a function. 
# It allows you to pass any number of positional arguments to the function, and they will be accessible
# as a tuple within the function.
```

KEYWORD ARGS - LIKE PAGELIST

```python
def print_details(**kwargs): ## **kwargs is used to pass a variable number of keyword arguments to a function. It allows you to pass any number of named arguments to the function, and they will be accessible as a dictionary within the function.
    for key, value in kwargs.items():
        print(f"{key}: {value}")
    
print_details(name="Alice", age=30, city="New York")
```

```python
def print_keyValueDetails(**args):
    for key, value in args.items(): ##its like pagelist.pxResults
        print(f"{key}:{value}")

print_keyValueDetails(name="Pavan", age=24, city="Hyderabad")
```

COMBINED ARGS - BOTH ARGS AND KEYWORD ARGS

```python
def combinedargs(*args, **kwargs):
    print("Positional arguments:")
    for arg in args:
        print(arg)
    
    print("\nKeyword arguments:")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
combinedargs(1, 2, 3, name="Alice", age=30, city="New York")
```

STRONG PASSWORD CHECKER

```python
def passwordCheck(password):
    if len(password)<8:
        return "password is too short"
    if not any(char.isupper() for char in password):
        return "password must contain at least one uppercase letter"
    if not any(char.islower() for char in password):
        return "password must contain at least one lowercase letter"
    if not any(char.isdigit() for char in password):
        return "password must contain at least one digit"
    if not any(char in "!@#$%^&*()_+-=[]{}|;:,.<>?/" for char in password):
        return "password must contain at least one special character"
    
    return "password is strong"

password = input("Enter a password: ")
result = passwordCheck(password)
print(result)
```

```python
def isStrongPassword(password):
    if len(password)<8:
        return "password is too short"
    if not any(char.isupper() for char in password):
        return "password must contain at least one uppercase letter"
    if not any(char.islower() for char in password):
        return "password must contain at least one lowercase letter"
    if not any(char.isdigit() for char in password):
        return "password must contain at least one digit"
    if not any(char in "!@#$%^&*()_+-=[]{}|;:,.<>?/" for char in password):
        return "password must contain at least one special character"
    
    return "password is strong"

password = input("Enter a password: ")
result = isStrongPassword(password)
print(result)
```
