```python
##For loop to find range of numbers. It always prints from 1 to 10 and the last number is skipped.
for i in range(1, 11):
    print(i)
```

```python
for i in range (10, 1, -1): ## Here we are decrementing the value.
    print(i)
```

```python
x= int(input("Enter a number: "))
sum=0
count=1
for i in range(x):
    sum=sum+count
    count+=1
print(sum)
```

```python
# reverse a string using for loop
string = "HelloWorld"
reversed_string = ""
for i in string:
    print(f"i = {i}")
    print(f"reversed_string = {reversed_string}")
    reversed_string = i + reversed_string
print(reversed_string)
```

```python
##Break
for i in range(1, 11):
    if i==6:
        break ## Break statement will exit the loop and control will be transferred to next statement after loop.
    print(i)
print(f"Loop ended at i={i}")
```

```python
##Pass statement
for i in range(1, 6):
    if i==3:
        pass ## Pass statement is a null statement. It is used when a statement is syntactically required but no action is needed.
    else:
        print("Loop completed successfully", i)
    print(i)  
```

```python
#Sum of first 10 natural numbers. 
n=10
sum=0
count=1
for i in range(n):
    sum=sum+count
    count+=1
    print(f"i:{i}, n:{n}")
print("The sum of first", n, "natural numbers is:", sum)
```

```python
##Factorial of a number
num = int(input("Enter a number to find factorial: "))  
factorial = 1
for i in range(1, num + 1):  
    factorial = factorial * i
print("The factorial of", num, "is", factorial)
```
