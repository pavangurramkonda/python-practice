```python
## If Statemnts and Loops
age =int(input("whats your age?"))
if age<13:
    print("you are a child")
elif age<18:
    print("You are minor")
else:
    print("you are adult")
```

```python
## nested if conditions
num = int(input("Enter a number: "))
if num >= 0:
    print("You entered a non-negative number")
    if num%2==0:
        print("The number is even")
    else:
        print("The number is odd")
else:
    print("You entered a negative number")
```

```python
year = int(input("Enter a year: "))
if year%4==0:
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

```python
fruits =["apple", "banana", "oranges"]
print(fruits)
```

```python
print(fruits[1]) ## Prints second element. 
print(fruits[1:]) ## Prints all from second element 
##fruits.append("Grapes")
##print(fruits)
fruits.remove("Grapes")
print(fruits)
```

```python
popped_fruits=fruits.pop() ## removes last element and returns it.
print(popped_fruits)
```

```python
index= fruits.index("oranges") ## returns index of element
print(index)
```

```python
fruits.insert(2, "oranges")
print(fruits.count("oranges"))
print(fruits)
```

```python
fruits[2]= "grapes"
fruits[3]="mango"
print(fruits)
for fruit in fruits:
    print(fruit)
```

```python
## nested if examples
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
operation = input("Enter operation (+, -, *, /))")
if operation == "+":
    result = num1+num2
    print("The sum is", result)
elif operation == "-":
    result = num1-num2
    print("The difference is", result)
elif operation == "*":
    result = num1*num2
    print("The product is", result)
else:
    if num2!=0:
         result = num1/num2
         print("The quotient is", result)
    else:
        print("Error: Division by zero is not allowed.")
```
